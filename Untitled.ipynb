```python
df.drop('group',axis=1,inplace=True)
```

```python
df['CryoSleep']=df['CryoSleep'].astype('bool')
df['VIP']=df['VIP'].astype('bool')
df['cab_num'] = df['cab_num'].astype('float')
```

### FEATURE ENGINNERING

```python
numerical_feat
```

```python
df.head()
```

```python
df1=pd.DataFrame(df[numerical_feat])
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1.drop(['CryoSleep','VIP','Transported'],axis=1,inplace=True)
```

```python
df1.columns
```

```python
from sklearn.impute import KNNImputer
imp= KNNImputer(n_neighbors=5) 
df1_imp= imp.fit_transform(df1)
df1 = pd.DataFrame(df1_imp, columns=df1.columns)
```

```python
df1.isnull().sum()
```

```python
cat_feat
```

```python
df2=pd.DataFrame(df[['CryoSleep','VIP','Transported']])
```

```python
df2.isnull().sum()
```

```python
df3=pd.DataFrame(df[cat_feat])
```

```python
df3.head()
```

```python
df3.isnull().sum()
```

```python
for feature in cat_feat:
    temp=df3[feature].mode()[0]
    df3[feature].fillna(temp,inplace=True)
```

```python
df3.isnull().sum()
```

```python
new_df=pd.concat([df1,df2,df3],axis=1)
```

```python
new_df.isnull().sum()
```

```python
### lets check relation between target and continuous variable
for feature in cont_feat:
    plt.Figure(figsize=(10,8))
    plt.scatter(df[feature],df['Transported'])
    plt.xlabel(feature)
    plt.ylabel('Transported')
    plt.show()
```

```python
new_df[cat_feat]
```

```python
new_df1=new_df.copy()
```

```python
new_df.isnull().sum()
```

```python
new_df1=pd.get_dummies(new_df1,columns=cat_feat,drop_first=True)
```

```python
new_df1.columns
```

```python
new_df1['CryoSleep'] = new_df1['CryoSleep'].astype(int)
new_df1['VIP'] = new_df1['VIP'].astype(int)
new_df1['Transported'] = new_df1['Transported'].astype(int)
```

```python
new_df1.head(5)
```

```python
new_df1['RoomFood']=new_df1['RoomService']*new_df1['FoodCourt']
new_df1['RoomShop']=new_df1['RoomService']*new_df1['ShoppingMall']
new_df1['RoomSpa']=new_df1['RoomService']*new_df1['Spa']
new_df1['RoomVR']=new_df1['RoomService']*new_df1['VRDeck']
```

```python
new_df1
```

### feature selection

```python
plt.figure(figsize=(20,12))
cor=new_df1.corr()
sns.heatmap(cor,annot=True,cmap=plt.cm.CMRmap_r)
plt.show()
```

```python
cont_feat.append('RoomFood')
cont_feat.append('RoomShop')
cont_feat.append('RoomSpa')
cont_feat.append('RoomVR')
```

```python
for feature in cont_feat:
    new_df1[feature]=np.log(new_df1[feature]+1)
```

```python
from sklearn.model_selection import train_test_split
X=new_df1.drop('Transported',axis=1)
Y=new_df1['Transported']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.02,random_state=0)
X_train.shape,X_test.shape
```

```python
from sklearn.feature_selection import mutual_info_classif
Mutual=mutual_info_classif(X_train,Y_train)
Mutual
```

```python
S=pd.Series(Mutual)
S.index=X_train.columns
S.sort_values(ascending=False)
```

```python
S.sort_values(ascending=False).plot.bar()
```

### after scaling

### neural network

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import RandomForestClassifier
forest=RandomForestClassifier(n_estimators=500,max_depth=12,min_samples_split=5,min_samples_leaf=3,max_features='log2')
clf=AdaBoostClassifier(random_state=70,base_estimator=forest,learning_rate=0.001)
clf.fit(X_train_scaled,Y_train)
```
